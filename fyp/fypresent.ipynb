```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

#é€æ˜Žãªã‚¹ã‚¿ã‚¤ãƒ«
plt.style.use('fivethirtyeight')

#TFPã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
A = 2
B = 1
C = 0.3

#ã‚³ãƒ–ãƒ€ã‚°ãƒ©ã‚¹ç”Ÿç”£é–¢æ•°ã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
alpha = 0.33

#ç”Ÿç”£é–¢æ•°ï¼ˆKã¨Lã®ã¿ï¼‰
def f(k, l):
    return k**(1/3) * l**(2/3)

#ã‚°ãƒªãƒƒãƒ‰ä½œæˆ
k = np.arange(0, 100, 2.5)
l = np.arange(0, 100, 2.5)
K, L = np.meshgrid(k, l)

#ï¼“ã¤ã®ç”Ÿç”£é–¢æ•°
Y1 = A*f(K, L)
Y2 = B*f(K, L)
Y3 = C*f(K, L)

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_wireframe(K, L, Y1, color='blue',linewidth=0.3) 
ax.plot_wireframe(K, L, Y2,linewidth=0.3)
ax.plot_wireframe(K, L, Y3, color='cyan',linewidth=0.3) 

# xè»¸ã®ç›®ç››è¨­å®š
ax.set_xticks([ 0, 50, 100])

# yè»¸ã®ç›®ç››è¨­å®š
ax.set_yticks([0, 50, 100])

# zè»¸ã®ç›®ç››è¨­å®š
ax.set_zticks([0, 50, 100, 150, 200])

#ãƒ©ãƒ™ãƒ«
ax.set_xlabel("K")
ax.set_ylabel("L")
ax.set_zlabel("Y")

ax.set_xlabel("K", size = 16, weight = "light")
ax.set_ylabel("L", size = 16, weight = "light")
ax.set_zlabel("Y", size = 16, weight = "light")
fig.subplots_adjust(bottom=0.1)

plt.tick_params(labelbottom=False,
                labelleft=False,
                labelright=False,
                labeltop=False)

#ãƒ•ã‚©ãƒ³ãƒˆã®è¨­å®š
plt.rcParams['font.family'] ='Arial'
plt.rcParams["font.size"] = 12

plt.show()

#ç”»åƒã®ä¿å­˜ã€dvi=è§£åƒåº¦
plt.savefig("fy10", transparent=True, dpi=600) # default: False
```
